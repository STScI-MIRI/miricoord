```python
import miricoord.miricoord.imager.mirim_tools as mt
import numpy as np
```

```python
xvec=np.arange(5,1027)
yvec=np.arange(0,1024)
```

```python
allx=np.zeros((len(yvec),len(xvec)))
ally=np.zeros((len(yvec),len(xvec)))
```

```python
for i in range(0,len(xvec)):
    ally[:,i]=yvec
    
for j in range(0,len(yvec)):
    allx[j,:]=xvec
```

```python
allx.shape
```

```python
from numpy import matlib as mb
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(2,2),dpi=250)
plt.title('yval')
plt.imshow(ally,origin='lower')
plt.colorbar(fraction=0.046, pad=0.1)
```

```python
plt.figure(figsize=(2,2),dpi=250)
plt.title('xval')
plt.imshow(allx,origin='lower')
plt.colorbar(fraction=0.046, pad=0.1)
```

```python
xlong=np.reshape(allx,len(xvec)*len(yvec))
ylong=np.reshape(ally,len(xvec)*len(yvec))
```

```python
v2,v3=mt.xytov2v3(xlong,ylong,'F770W')
```

```python
xlong2,ylong2=mt.v2v3toxy(v2,v3,'F770W')
```

```python
allx2=np.reshape(xlong2,(len(yvec),len(xvec)))
ally2=np.reshape(ylong2,(len(yvec),len(xvec)))
xdiff=allx-allx2
ydiff=ally-ally2
```

```python
plt.figure(figsize=(2,2),dpi=250)
plt.title('xdiff')
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
plt.imshow(xdiff,origin='lower')
plt.colorbar(fraction=0.046, pad=0.1)
```

```python
plt.figure(figsize=(2,2),dpi=250)
plt.title('ydiff')
plt.xlabel('X pixel')
plt.ylabel('Y pixel')
plt.imshow(ydiff,origin='lower')
plt.colorbar(fraction=0.046, pad=0.1)
```
